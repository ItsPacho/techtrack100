```python
# imports 
import pandas as pd
```

```python
table = pd.read_html('http://www.fasttrack.co.uk/league-tables/tech-track-100/league-table/' , header=0, index_col=0)[0]
table.head()
```

```python
# cleaning the data

# removing unwanted symbols in the Latest sales column
table['Latest sales £000s'] = table['Latest sales £000s'].str.replace('*', '').str.replace('†', '').str.replace(',','')
```

```python
# splitting the company name from the description
companies = table['Company']

# function for reversing the string, splitting it on the first uppercase letter, reversing it into the original order
# and then trimming the leading whitespace of the string (if there is one) 

def splitter(company):
    reverse_string = company[::-1]
    
    for char in reverse_string:
        if char.isupper():
            reverse_string = reverse_string.replace(char, char+" ")
            break
    return reverse_string[::-1].lstrip().split(" ", 1)

companies_cleaned = []
for company in companies:
    companies_cleaned.append(splitter(company))

# and finally splitting the string into a list of company name and description

Company = []
Description = []
for i, el in enumerate(companies_cleaned):
    Company.append(companies_cleaned[i][0])
    Description.append(companies_cleaned[i][1])
```

```python
# inserting the values into the table
table['Company'] = Company
table.insert(1, 'Description', Description)
```
